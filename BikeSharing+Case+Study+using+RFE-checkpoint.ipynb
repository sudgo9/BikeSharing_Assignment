## Model Selection using RFE (Bike Sharing)

### Importing and Understanding Data

```python
# Supress Warnings

import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm  
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import r2_score

df = pd.read_csv('C:\\11 WorkStuff\\Persona\\Learnings\\Bike_Sharing_Modelling\\day.csv')


# List of variables to map

varlist =  ['weathersit']

# Defining the map function
def binary_map(x):
    return x.map({1: 'Clear', 2: 'Misty',3:'Light Rain',4:'Heavy Rain'})

# Applying the function to the df list
df[varlist] = df[varlist].apply(binary_map)

# List of variables to map

varlist2 =  ['season']

# Defining the map function
def binary_map2(x):
    return x.map({1: 'Spring', 2: 'Summer',3:'Fall',4:'Winter'})

# Applying the function to the df list
df[varlist2] = df[varlist2].apply(binary_map2)

df.drop(['dteday','instant'], axis = 1, inplace = True)


# plot graphs to review data

plt.figure(figsize = (10, 5))
sns.boxplot(x = 'mnth', y = 'cnt', hue = 'weathersit', data = df)

plt.figure(figsize = (10, 5))
sns.boxplot(x = 'yr', y = 'cnt', hue = 'mnth', data = df)

sns.catplot(x = "mnth",
            y = "cnt",
            hue = "yr",
            col = "weathersit",
            data = df,
            kind = "violin",
            split = True)

sns.catplot(x = "mnth",
            y = "cnt",
            hue = "yr",
            col = "season",
            data = df,
            kind = "violin",
            split = True)


plt.figure(figsize = (16, 10))
sns.heatmap(df.corr(), annot = True, cmap="YlGnBu")

plt.subplot(2,3,2)
sns.boxplot(x = 'yr', y = 'cnt', data = df)
plt.subplot(2,3,3)
sns.boxplot(x = 'mnth', y = 'cnt', data = df)
plt.subplot(2,3,4)
sns.boxplot(x = 'holiday',  y = 'cnt', data = df)
plt.subplot(2,3,5)
sns.boxplot(x = 'weekday',  y = 'cnt', data = df)
plt.subplot(2,3,6)
sns.boxplot(x = 'workingday', y = 'cnt', data = df)

plt.subplot(2,3,5)
sns.boxplot(x = 'season', y = 'cnt', data = df)

plt.subplot(2,3,6)
sns.boxplot(x = 'weathersit', y = 'cnt', data = df)

plt.show()



seasons = pd.get_dummies(df['season'])
# # Let's drop the first column from status df using 'drop_first = True'
seasons = pd.get_dummies(df['season'], drop_first = True)
# # Add the results to the original housing dataframe
df = pd.concat([df, seasons], axis = 1)


weathersits = pd.get_dummies(df['weathersit'])
# # Let's drop the first column from status df using 'drop_first = True'
weathersits = pd.get_dummies(df['weathersit'], drop_first = True)
# # Add the results to the original housing dataframe
df = pd.concat([df, weathersits], axis = 1)

df.drop(['weathersit','season'], axis = 1, inplace = True)

# We specify this so that the train and test data set always have the same rows, respectively

df_train, df_test = train_test_split(df, train_size = 0.6, test_size = 0.4, random_state = 100)



#from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_vars = ['temp', 'hum', 'windspeed', 'cnt']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])


y_train = df_train.pop('cnt')
X_train = df_train



# # # Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm, 20)             # running RFE
rfe = rfe.fit(X_train, y_train)

list(zip(X_train.columns,rfe.support_,rfe.ranking_))

col = X_train.columns[rfe.support_]
col

X_train.columns[~rfe.support_]

# # Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]

# Adding a constant variable 
X_train_rfe = sm.add_constant(X_train_rfe)
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
#Let's see the summary of our linear model
print(lm.summary())




# Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['Spring'], axis = 1)
# Build a third fitted model
X_train_lm2 = sm.add_constant(X_train_rfe)
lr_2 = sm.OLS(y_train, X_train_lm2).fit()
# Print the summary of the model
#print(lr_2.summary())



# # Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['mnth'], axis = 1)
# Build a third fitted model
X_train_lm3 = sm.add_constant(X_train_rfe)
lr_3 = sm.OLS(y_train, X_train_lm3).fit()
# Print the summary of the model
#print(lr_3.summary())




# # # Dropping insignificant variables

# # Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['Light Rain'], axis = 1)
# Build a third fitted model
X_train_lm4 = sm.add_constant(X_train_rfe)
lr_4 = sm.OLS(y_train, X_train_lm4).fit()
# Print the summary of the model
#print(lr_4.summary())



# # Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['temp'], axis = 1)
# Build a third fitted model
X_train_lm5 = sm.add_constant(X_train_rfe)
lr_5 = sm.OLS(y_train, X_train_lm5).fit()
# Print the summary of the model
#print(lr_5.summary())


# # Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['holiday'], axis = 1)
# Build a third fitted model
X_train_lm6 = sm.add_constant(X_train_rfe)
lr_6 = sm.OLS(y_train, X_train_lm6).fit()
# Print the summary of the model
#print(lr_6.summary())



# # Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['yr'], axis = 1)
# Build a third fitted model
X_train_lm7 = sm.add_constant(X_train_rfe)
lr_7 = sm.OLS(y_train, X_train_lm7).fit()
# Print the summary of the model
#print(lr_7.summary())


# # Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['workingday'], axis = 1)
# Build a third fitted m
X_train_lm8 = sm.add_constant(X_train_rfe)
lr_8 = sm.OLS(y_train, X_train_lm8).fit()
# Print the summary of the model
#print(lr_8.summary())

# # Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['hum'], axis = 1)
# Build a third fitted model
X_train_lm9 = sm.add_constant(X_train_rfe)
lr_9 = sm.OLS(y_train, X_train_lm9).fit()
# Print the summary of the model
#print(lr_9.summary())


# # Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['Winter'], axis = 1)
# Build a third fitted model
X_train_lm10 = sm.add_constant(X_train_rfe)
lr_10 = sm.OLS(y_train, X_train_lm10).fit()
# Print the summary of the model
#print(lr_10.summary())

# # Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['windspeed'], axis = 1)
# Build a third fitted model
X_train_lm11 = sm.add_constant(X_train_rfe)
lr_11 = sm.OLS(y_train, X_train_lm11).fit()
# Print the summary of the model
#print(lr_11.summary())

# # Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['atemp'], axis = 1)
# Build a third fitted model
X_train_lm12 = sm.add_constant(X_train_rfe)
lr_12 = sm.OLS(y_train, X_train_lm12).fit()
# Print the summary of the model
#print(lr_12.summary())

# Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['Summer'], axis = 1)
# Build a third fitted model
X_train_lm14 = sm.add_constant(X_train_rfe)
lr_14 = sm.OLS(y_train, X_train_lm14).fit()
# Print the summary of the model
#print(lr_14.summary())

# Dropping insignificant variables
X_train_rfe = X_train_rfe.drop(['Misty'], axis = 1)
# Build a third fitted model
X_train_lm15 = sm.add_constant(X_train_rfe)
lr_15 = sm.OLS(y_train, X_train_lm15).fit()
# Print the summary of the model
print(lr_15.summary())



# Calculate the VIFs for new model
X_train_rfe1 = X_train_rfe.drop(["const"], axis = 1)
vif = pd.DataFrame()
X = X_train_rfe1
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif

# Build a third fitted model
X_train_lm = sm.add_constant(X)
lr_16 = sm.OLS(y_train, X_train_lm).fit()
# Print the summary of the model
print(lr_16.summary())
y_train_pred = lr_16.predict(X_train_lm)

 
#Predict values  
y_train_pred_rfe = lr_15.predict(X_train_rfe)
res=y_train - y_train_pred

#Checking normality:

# fig = plt.figure()
# sns.distplot((res), bins = 20)
# fig.suptitle('Error Terms', fontsize = 20)                 
# plt.xlabel('Errors', fontsize = 18) 

num_vars = ['temp', 'hum', 'windspeed', 'cnt']
df_test[num_vars] = scaler.transform(df_test[num_vars])
y_test = df_test.pop('cnt')
X_test = df_test
y_test_rfe = y_test 
X_test_rfe = X_test


X_test = sm.add_constant(X_test)
X_test_rfe = sm.add_constant(X_test_rfe)

# predicting using values used by the final model
test_col = X_train_lm.columns
X_test=X_test[test_col[1:]]
# Adding constant variable to test dataframe
X_test = sm.add_constant(X_test)
X_test.info()
X_train_rfe.columns # RFE Model
X_test.columns


test_col1 = X_train_rfe.columns
X_test_rfe=X_test_rfe[test_col1[1:]]
# Adding constant variable to test dataframe
X_test1 = sm.add_constant(X_test_rfe)
X_test_rfe.info()

X_train_rfe.columns # RFE Model
X_test.columns

y_pred = lr_16.predict(X_test)

r2_score(y_test, y_pred)


y_pred_rfe = lr_16.predict(X_test1)
r2_score(y_test, y_pred_rfe)


# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_lm.columns]

# Adding a constant variable 
X_test_new1 = sm.add_constant(X_test_new)
X_test_new1.head()

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
mse
mse = mean_squared_error(y_test, y_pred_rfe)
mse

# Plotting y_test and y_pred to understand the spread - Normal Model
fig = plt.figure()
plt.scatter(y_test, y_pred)
fig.suptitle('y_test vs y_pred', fontsize = 20)              # Plot heading 
plt.xlabel('y_test', fontsize = 18)                          # X-label
plt.ylabel('y_pred', fontsize = 16) 




#X_train.head(500)
```
